```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Loading the Iris dataset
iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)

data.head()
```

```python
#KMeans Clustering
from sklearn.cluster import KMeans

# Applying KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
data['kmeans_cluster'] = kmeans.fit_predict(data)


# Visualizing the KMeans clusters 
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data.iloc[:, 0], y=data.iloc[:, 1], hue=data['kmeans_cluster'], palette='viridis', s=100, edgecolor='black')
plt.title("KMeans Clustering on Iris Dataset")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

```python
#Hierarchical Clustering

from sklearn.cluster import AgglomerativeClustering

# Applying Hierarchical clustering
hierarchical = AgglomerativeClustering(n_clusters=3)
data['hierarchical_cluster'] = hierarchical.fit_predict(data)

# Visualizing the Hierarchical clusters 
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data.iloc[:, 0], y=data.iloc[:, 1], hue=data['hierarchical_cluster'], palette='viridis', s=100, edgecolor='black')
plt.title("Hierarchical Clustering on Iris Dataset")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```
